```python
import os
import chromadb
```

```python
documents_directory = "documents/SS123"
collection_name = "SS123"
persist_directory = documents_directory
```

```python
client = chromadb.PersistentClient(path=persist_directory)
collection = client.get_or_create_collection(name=collection_name)
```

```python
print(collection.count())
```

```python
sections = list()
data = list()

files = os.listdir(documents_directory)
documentExtension = "txt"

files = [ file for file in files if file.endswith(documentExtension) ]

for filename in files:
    with open(f"{documents_directory}/{filename}") as file:
        for line_number, line in enumerate(file.readlines()):
            if "---" in line:
                # print(line_number)
                # print(line)
                sections.append(line_number)
            
            data.append(line)

print(sections)
print(data)
```

```python
metadatas = list()
documents = list()

for i in range(len(sections)): 
    
    to_clean = data[sections[i]+1]
    cleaned_data = to_clean.strip()

    # metadatas.append( {"Source": data[sections[i]+1]} )
    metadatas.append( {"Source": cleaned_data } )

    if i < len(sections)-1:
        sectionData = data[sections[i]+1+1 : sections[i+1]]
    else:
        sectionData = data[ sections[i]+1+1 :  ]    
        
    sectionData = [''.join(sectionData)]
    sectionData = sectionData[0].strip()

    documents.append(sectionData)

ids = [str(i) for i in range(0, len(documents))]
```

```python
print(metadatas)
print(documents)
print(ids)
```

```python
for i in range(65):
    print(ids[i])
    print(metadatas[i])
    print(documents[i])
    print("\n")
```

```python
for i in range(0, len(documents), 5):
        collection.add( 
            ids=ids[i:i+5],
            documents=documents[i:i+5],
            metadatas=metadatas[i:i+5],
        )
```
